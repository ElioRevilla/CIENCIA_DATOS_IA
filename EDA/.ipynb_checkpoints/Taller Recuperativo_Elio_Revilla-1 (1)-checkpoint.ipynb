Attribute documentation:

**Variables Categoricas:**

- 3 age: age in years
- 4 sex: sex (1 = male; 0 = female)
- 9 cp: chest pain type 
 - Value 1: typical angina
 - Value 2: atypical angina
 - Value 3: non-anginal pain
 - Value 4: asymptomatic
- 16 fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
- 19 restecg: resting electrocardiographic results
 - Value 0: normal
 - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
 - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
- 38 exang: exercise induced angina (1 = yes; 0 = no)
- 41 slope: the slope of the peak exercise ST segment  (The ST Segment represents the interval between ventricular depolarization and repolarization)
 - Value 1: upsloping
 - Value 2: flat
 - Value 3: downsloping
- 44 ca: number of major vessels (0-3) colored by flourosopy
- 51 thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
- 58 num: diagnosis of heart disease (angiographic disease status)
 - Value 0: < 50% diameter narrowing
 - Value 1: > 50% diameter narrowing
 
**Variables No Categoricas:**
- 10 trestbps: resting blood pressure (in mm Hg on admission to the hospital)
- 12 chol: serum cholestoral in mg/dl
- 32 thalach: maximum heart rate achieved
- 40 oldpeak = ST depression induced by exercise relative to rest

#### 3. ¿Qué variables cardiológicas se ven más afectadas por la edad?

Para determinar qué variables cardiológicas se ven más afectadas por la edad, podemos calcular la correlación entre la edad y cada una de las variables cardiológicas en el conjunto de datos. La correlación te dará una medida de la relación lineal entre la edad y cada variable.

```python
# Al intentar realizar la matriz de correlacion se tuvo problemas convertir string a float, puede que sean
# datos faltantes o nulos, para solucionar esto aplicamos el siguiente comando
clev1 = pd.read_csv('processed.cleveland.data', names=names, na_values='?')
```

```python
# Cálculo de la matriz de correlación
correlation_matrix = clev1.corr()

# Visualización de la matriz de correlación en un mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de correlación entre variables cardiológicas')
plt.show()
```

```python
# Cálculo de las correlaciones entre la edad y las variables cardiológicas
age_correlations = clev1.corr()['age'].sort_values(ascending=False)

# Mostrar las correlaciones
print(age_correlations)
```

La variable "ca" (número de vasos principales) muestra la correlación más fuerte con la edad, con un valor de correlación positiva de 0.362605. Esto sugiere que a medida que aumenta la edad, es más probable que se observe un mayor número de vasos principales con alguna obstrucción o enfermedad.

Otras variables que muestran una correlación moderada con la edad son "trestbps" (presión arterial en reposo), "num" (diagnóstico de enfermedad cardíaca), "chol" (nivel de colesterol en suero), "oldpeak" (depresión del segmento ST inducida por el ejercicio) y "slope" (pendiente del segmento ST en el pico del ejercicio). Estas variables tienen valores de correlación positiva entre 0.20 y 0.28, lo que indica una tendencia de aumento en su medida a medida que aumenta la edad.

Por otro lado, la variable "thalach" (frecuencia cardíaca máxima alcanzada) muestra una correlación negativa fuerte con la edad, con un valor de correlación de -0.393806. Esto sugiere que a medida que aumenta la edad, es probable que la frecuencia cardíaca máxima disminuya.

Otras variables, como "restecg" (resultados electrocardiográficos en reposo), "thal" (resultado de la prueba de talio), "fbs" (nivel de azúcar en sangre en ayunas), "cp" (tipo de dolor en el pecho) y "exang" (angina inducida por ejercicio), tienen correlaciones más débiles con la edad, pero aún así muestran cierta asociación.

#### 4. ¿Qué variables muestran mayor correlación?

Correlacion positiva

```python
# Cálculo de la matriz de correlación
correlation_matrix = clev1.corr()

# Obtención de los pares de variables con mayor correlación, con valores menores a 1
highly_correlated_pairs = correlation_matrix.unstack().sort_values(ascending=False)
highly_correlated_pairs = highly_correlated_pairs[highly_correlated_pairs < 1]  
# Mostrar los pares de variables y sus correlaciones más altas
print(highly_correlated_pairs.head(5))
```

Oldpeak" y "slope" tienen una correlación de 0.577537. Esto indica una relación positiva moderada entre la depresión del segmento ST inducida por el ejercicio y la pendiente del segmento ST en el pico del ejercicio. Es decir, a medida que la depresión del segmento ST aumenta, es más probable que la pendiente del segmento ST también aumente.

"ca" y "num" tienen una correlación de 0.518909. Esto sugiere una relación positiva moderada.

"thal" y "num" tienen una correlación de 0.509923. Esto indica una relación positiva moderada, es decir, un resultado anormal en la prueba está asociado con un mayor diagnóstico de enfermedad cardíaca.

Correlación negativa

```python
# Filtrar los valores de correlación negativa
negative_correlations = correlation_matrix.unstack().sort_values(ascending=True)
negative_correlations = negative_correlations[negative_correlations < 0]  # Filtrar valores menores a 0

# Mostrar los pares de variables y sus correlaciones más negativas
print(negative_correlations.head(5))
```

thalach" y "num" tienen una correlación de -0.415040. Esto sugiere que a medida que la frecuencia cardíaca máxima alcanzada disminuye, es más probable que el diagnóstico de enfermedad cardíaca sea más alto. Se podría decir, una menor frecuencia cardíaca máxima está asociada con un mayor riesgo de enfermedad cardíaca.

"age" y "thalach" tienen una correlación de -0.393806. Esto indica que a medida que la edad aumenta, es más probable que la frecuencia cardíaca máxima alcanzada disminuya. Edad está inversamente relacionada con la frecuencia cardíaca máxima.

"thalach" y "slope" tienen una correlación de -0.385601. Esto sugiere que a medida que la frecuencia cardíaca máxima alcanzada disminuye, es más probable que la pendiente del segmento ST en el pico del ejercicio sea más pronunciada hacia abajo. una menor frecuencia cardíaca máxima está asociada con una pendiente más pronunciada hacia abajo en el segmento ST durante el ejercicio.

### 5. ¿Qué variables son determinantes para predecir patologías cardíacas?

```python
num_correlations = clev1.corr()['num'].sort_values(ascending=False)

# Mostrar las correlaciones
print(num_correlations)
```

Las variables "ca", "thal", "oldpeak" y "cp" muestran las correlaciones más altas con la presencia de enfermedad cardíaca (columna "num"). La variable "thalach" muestra una correlación negativa con la presencia de enfermedad cardíaca. Esto significa que las variables "ca", "thal", "oldpeak", "cp" y "thalach" pueden ser consideradas como variables determinantes para predecir patologías cardíacas. 

**Distribución variable vs  num**

```python
# Según dataset la columna num contiene valores desde 0 a 4, donde 0 representa la ausencia de la enfermedad y los 
# valores 1,2,3 y 4, la presencia de esta.
# Reemplazaremos las variables 2,3,4 por 1, que va simbolizar la presencia la presencia de la enfermedad.
clev2 = clev1['num'].replace([2, 3, 4], 1)
print(clev2.value_counts())

fig, ax = plt.subplots(figsize=(5,4))
name = ["No Enfermedad: 0", "Enfermedad: 1"]
ax = clev2.value_counts().plot(kind='bar')
ax.set_title("Enfermedad", fontsize = 13, weight = 'bold')
ax.set_xticklabels (name, rotation = 0)

# To calculate the percentage
totals = []
for i in ax.patches:
    totals.append(i.get_height())
total = sum(totals)
for i in ax.patches:
    ax.text(i.get_x()+.09, i.get_height()-50, \
            str(round((i.get_height()/total)*100, 2))+'%', fontsize=14,
                color='white', weight = 'bold')
    
plt.tight_layout()
```

Hay menos enfermos en el dataset.
