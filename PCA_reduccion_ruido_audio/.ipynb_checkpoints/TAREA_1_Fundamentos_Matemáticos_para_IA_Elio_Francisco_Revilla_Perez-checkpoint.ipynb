**Aplicacion de algoritmo k-means**

Este problema trataremos el problema de clustering, es decir, determinar los k representates de un conjunto de datos y las etiquetas de estos que minimizan la distancia de cada cluster a su representante. <br>
Programe el algortimo de k-means. Construya una rutina <br>
                         <center>k_means_fit(X, Z0, NITERMAX)</center> <br>
donde X son los datos Z0 son los representates iniciales, y NITERMAX es el número máximo de iteraciones. <br>


```python
def rep_inicial(K, X, semilla=0):
    """Genera los representantes inciales en base a una seleccion aleatoria desde los datos. 
    Selecciona una semilla para reproducibilidad.
    También devuelve el K para utilizarlo en las siguientes funciones.  """
   
    np.random.seed(semilla)
    idx = np.random.permutation(X.shape[0])[:K]
    representantes = X[idx]
    return representantes, K

def vecino_cercano(x, representantes, K):
    """Devuelve el índice del vecino más cercano de un vector x de los datos respecto de cada representante."""
    distancias = np.empty(K)
    for i in range(K):
        distancias[i] = np.linalg.norm(representantes[i] - x)
    return np.argmin(distancias)

def crear_clusters(X, K ,representantes):
    """Crea clusters con los índices del vecino más cercano."""
    m = np.shape(X)[0]
    cluster_idx = np.empty(m)   
    for i in range(m):
        cluster_idx[i]= vecino_cercano(X[i], representantes, K)
    return cluster_idx

def calcular_representantes(X, K, cluster_idx):
    """Toma un cluster con índices, selecciona de los datos los elementos del cluster y los promedia
    para devolver los nuevos representantes.   
    """
    _, n = np.shape(X)
    representantes = np.empty((K, n))
    for i in range(K):
        j = X[cluster_idx == i]              
        representantes[i] = np.mean(j, axis=0)
    return representantes

def calcula_jclust(X, clusters_idx, representantes):
    m = np.shape(X)[0]
    jclust = np.empty(m)
    for i in range(m):
        jclust[i] = np.linalg.norm(X[i] - representantes[np.int32(clusters_idx[i])])
    return np.divide(np.sum(jclust), m)

def k_means(X, Z0, NITERMAX):
    """Junta todas las funciones anteriores y las itera hasta que se cumpla la condición de que el nuevo representante 
    sea igual al previo  o hasta que se cumpla el máximo de iteraciones.
    """
    Zn = Z0[0]
    K = Z0[1]
    jclust = np.empty((0, 1))
    for niter in range(NITERMAX):        
        clusters= crear_clusters(X, K, Zn)
        jclust_it = calcula_jclust(X, clusters, Zn)
        jclust = np.append(jclust, jclust_it)
        Z = calcular_representantes(X, K, clusters)
       
        if np.array_equal(Z, Zn):
            return clusters, Z, jclust
        else:
            Zn = Z
  
    return clusters, Z, jclust

def k_means_fit(X, Z0, NITERMAX):
  Z0_new = rep_inicial(Z0, X, semilla=0)
  clusters, rep, jclust = k_means(X, Z0_new, NITERMAX)
  return clusters, rep, jclust
```

```python
 clusters, rep, jclust = k_means_fit(X, Z0, NITERMAX)
```

d) Discuta porque el algoritmo de k − means converge, es decir, es seguro que de una iteración a la siguiente la función objetivo decrece?

La convergencia del algoritmo K-means se da porque se tiene la premisa  de minimización de una función objetivo,  en cada iteración, se busca minimizar la suma de cuadrados de las distancias entre los puntos y sus centroides, como resultado, el valor de la función objetivo disminuirá o permanecerá igual en cada iteración.

3. (5 puntos) **Independencia Lineal**

Considere una matriz de $A \in R^{mxn}$ con $m>n$ y un vector $b\in R^{m} $. Buscamos el vector solución $x \in R^{n}$ del sistema lineal $ Ax=b$. Describa como puede asegurar de forma práctica si la solución del sistema existe o no.

**SOLUCION**

**El sistema no tiene solución:** Esto sucede cuando las filas de $A$ son linealmente dependientes, lo que significa que una o más filas pueden expresarse como una combinación lineal de las demás filas. En este caso, al intentar resolver el sistema, se llega a una contradicción o a una ecuación sin solución. Para asegurarse de que el sistema no tenga solución, se puede calcular el rango de la matriz $A$ y compararlo con el rango de la matriz aumentada $[A|b]$. Si el rango de $A$ es menor que el rango de $[A|b]$, entonces el sistema no tiene solución.

**El sistema tiene solución:** Esto sucede cuando las filas de $A$ son linealmente independientes, lo que significa que cada fila aporta información adicional y no se puede expresar como combinación lineal de las demás filas. En este caso, al intentar resolver el sistema, se llega a una solución única o a un conjunto de soluciones. Para asegurarse de que el sistema tenga solución, se puede calcular el rango de la matriz $A$ y compararlo con el rango de la matriz aumentada $[A|b]$. Si el rango de $A$ es igual al rango de $[A|b]$, entonces el sistema tiene solución. Podemos utilizar métodos como la eliminación gaussiana o la factorización QR para encontrar la solución del sistema.

4. (15 puntos) **Clasificador binario: Perceptrón**

En este problema trataremos el problema de clasificación binaria, es decir, determinar si un dato pertenece a una de dos clases. 

a) Programe el algoritmo del Perceptrón. Puede ser útil construir en este caso una rutina.

                          $Perceptron_fit(X,y,nitmax,eat)$

que tome los datos $X$, sus etiquetas y, y que retorne los pesos w del clasificador. Otro hiperparametro a considerar son nitmax y era que corresponden respectivamente al número máximo de iteraciones del algoritmo y el *learning rate*

```python
#función escalón unitario
def step_func(z):
        return 1.0 if (z > 0) else 0.0
    
def perceptron_fit(X, y, epochs, lr):   
   
    m, n = X.shape
    
    # Initializing parapeters(theta) to zeros.
    # +1 in n+1 for the bias term.
    theta = np.zeros((n+1,1))
    
    # Empty list to store how many examples were 
    # misclassified at every iteration.
    n_miss_list = []
    
    # Training.
    for epoch in range(epochs):
        
        # variable to store #misclassified.
        n_miss = 0
        
        # looping for every example.
        for idx, x_i in enumerate(X):
            
            # Insering 1 for bias, X0 = 1.
            x_i = np.insert(x_i, 0, 1).reshape(-1,1)
            
            # Calculating prediction/hypothesis.
            y_hat = step_func(np.dot(x_i.T, theta))
            
            # Updating if the example is misclassified.
            if (np.squeeze(y_hat) - y[idx]) != 0:
                theta += lr*((y[idx] - y_hat)*x_i)
                
                # Incrementing by 1.
                n_miss += 1
        
        # Appending number of misclassified examples
        # at every iteration.
        n_miss_list.append(n_miss)
        
    return theta


def plot_decision_boundary(X, theta, y):    
    
    x1 = [min(X[:,0]), max(X[:,0])]
    m = -theta[1]/theta[2]
    c = -theta[0]/theta[2]
    x2 = m*x1 + c
    
    # Plotting
    fig = plt.figure(figsize=(10,8))
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "r^")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs")
    plt.xlabel("feature 1")
    plt.ylabel("feature 2")
    plt.title("Perceptron Algorithm")
    plt.plot(x1, x2, 'y-')
    
```

```python
w = perceptron_fit(X, y, nitmax, eat)
```

b) Programe una rutina que dado un vector de pesos *w*, y un conjunto de datos X prediga en a que conjunto de datos pertence cada dato.

```python
import numpy as np

def perceptron_predict(X, w):
    # Agregar columna de unos a X para tener en cuenta el bias
    X = np.c_[np.ones((X.shape[0], 1)), X]
    
    # Calcular la salida del perceptrón para cada dato en X ,  función escalón unitario
    y_pred =  np.heaviside(np.dot(X, w),0)
    
    return y_pred

y_pred = perceptron_predict(X, w)

 
```

c) Programe una rutina que dado un vector de pesos *w*, un conjunto de datos X y sus etiquetas y, entregue un score de que tan bien clasificados están los datos según las predicciones hechas por el clasificador.

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_fscore_support

def perceptron_metrics(X, y, w):
    # Agregar columna de unos a X para tener en cuenta el bias
    X = np.c_[np.ones((X.shape[0], 1)), X]
    
    # Calcular la salida del perceptrón para cada dato en X
    y_pred = np.heaviside(np.dot(X, w),0)
    
    # Calcular la matriz de confusión
    tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()
    
    # Calcular la tasa de aciertos
    score = accuracy_score(y, y_pred)
    
    # Calcular la precisión, exhaustividad y puntuación F1
    precision, recall, f1, _ = precision_recall_fscore_support(y, y_pred, average='binary')
    
    return score, tn, fp, fn, tp, precision, recall, f1

# Calcular las métricas del perceptrón
score, tn, fp, fn, tp, precision, recall, f1 = perceptron_metrics(X, y, theta)

```

a) Calcule la despomposición en valores singular de la matriz de asociada a los datos utilizando la funcion **svd** de la librería **numpy.linalg**

```python
from tensorflow import keras
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
X = x_test
X= X.reshape(*X.shape[:-2], -1)
```

```python
U, sigma, V = np.linalg.svd(X, full_matrices=False)
```

```python
# Mostrar las dimensiones de las matrices resultantes
print("Dimensiones de U:", U.shape)
print("Dimensiones de s:", sigma.shape)
print("Dimensiones de Vt:", V.shape)
```

La matriz U contiene los vectores singulares izquierdos, sigma contiene los valores singulares y V contiene los vectores singulares derechos. 

```python

import matplotlib.pyplot as plt

# Seleccionar el número de componentes principales para la reconstrucción
n_components = 40

# Reconstruir las imágenes utilizando los componentes principales
X_reconstructed = U[:, :n_components] @ np.diag(sigma[:n_components]) @ V[:n_components, :]

# Visualizar las primeras 10 imágenes originales y reconstruidas
fig, axs = plt.subplots(10, 2, figsize=(10, 20))
for i in range(10):
    axs[i, 0].imshow(X[i].reshape(28,28), cmap='gray')
    axs[i, 1].imshow(X_reconstructed[i].reshape(28, 28), cmap='gray')
    axs[i, 0].set_title("Original")
    axs[i, 1].set_title("Reconstruida")
plt.tight_layout()
plt.show()
```
